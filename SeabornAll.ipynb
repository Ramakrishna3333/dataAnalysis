```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds=sns.load_dataset("diamonds")
print(diamonds)
sns.boxplot(x="carat",y="price",data=diamonds)
plt.title("box plot of carat vs price")
plt.xlabel("carat")
plt.xlabel("price($)")
plt.show()
```
